```python
import inspect
import warnings
from typing import List, Optional, Union

import torch
from torch import autocast
from tqdm.auto import tqdm

```

```python
from diffusers import StableDiffusionImg2ImgPipeline
```

```python
!pip install transformers scipy ftfy accelerate
```

```python
device = "cuda"
model_path = "CompVis/stable-diffusion-v1-4"

pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
    model_path,
    revision="fp16",
    torch_dtype=torch.float16,
    use_auth_token=True
)
pipe = pipe.to(device)
```

```python
import requests
from io import BytesIO
from PIL import Image

# url = "https://www.google.com/url?sa=i&url=https%3A%2F%2Fsteemit.com%2Fart%2F%40hyojunguy%2Felon-musk-sketch&psig=AOvVaw1jbc9WI0dVaZnX9JuGsHBh&ust=1704135450034000&source=images&cd=vfe&opi=89978449&ved=0CBIQjRxqFwoTCLit0butuoMDFQAAAAAdAAAAABAD"
path = "/content/drive/MyDrive/AI App/0dcfb548989afdf22afff75e2a46a508.jpg"
# response = requests.get(path)
# init_img = Image.open(BytesIO(response.content)).convert("RGB")

init_img = Image.open(path)
init_img = init_img.resize((768, 512))
init_img
```

```python
prompt = "Created a different concept of this eagle logo"
```

```python
generator = torch.Generator(device=device).manual_seed(1024)
with autocast("cuda"):
    image = pipe(prompt=prompt, init_image=init_img, strength=0.9, guidance_scale=7.5, generator=generator).images[0]
```
